## What did you know about your dataset?


We can see our dataset contains 48,895 rows/entries and 16 columns/attributes. Let's delve into the details of these variables:

* id: A unique identifier for an Airbnb listing.
* name: The title or name of the listed property or room type on the platform.
*host_id: A unique identifier for an Airbnb host.
*host_name: The name of the person under whom the host is registered.
*neighbourhood_group: A grouping of areas.
*neighbourhood: The specific area falling under the neighbourhood_group.
*latitude: The latitude coordinate of the listing.
*longitude: The longitude coordinate of the listing.
*room_type: A categorization of the type of listing rooms.
*price: The price of the listing.
*minimum_nights: The minimum number of nights required for a single visit.
*number_of_reviews: The total count of reviews provided by visitors.
*last_review: The content of the last review given.
*reviews_per_month: An indicator of reviews received per month, highlighting the activity of the listing.
*calculated_host_listings_count: The total number of listings registered under the host.
*availability_365: The number of days in a year for which a host makes their property available.

####**Chart - 1**

```python
# chart 1 visulization code
# calculate the average price of neighbourhood group
avg_price = airbnb_df.groupby(['neighbourhood_group'])['price'].mean().reset_index()

# create a bar plot using seaborn

plt.figure(figsize = (8, 6))
prefer = sns.barplot(data = avg_price, x = 'neighbourhood_group',y ='price', palette='dark')
lebel = prefer.bar_label(prefer.containers[0])
plt.xlabel('Neighbourhood_group',fontsize = 14)
plt.ylabel('Average Price',fontsize = 14)
plt.title('Average Price in differant Neighbourhood Group',fontsize = 16)
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 12)
plt.show()
```

####**Chart - 9** - Heatmap

```python
# Chart - 9      Visulization code

# Calculate the correlation matrix using Kendall method
corr_matrix = airbnb_df.corr(method='kendall')

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap with annotations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Set the title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

##### 1. Why did you pick the specific chart?

Through the Correlation Heatmap, we can observe the interplay and associations between numerical attributes in the DataFrame, all represented by varying color intensities.

##### 2. What is/are the insight(s) found from the chart?

In the Heatmap Matrix some variables with some other have positive correlation and some also have negative correlation. e.g.

Positive corr : Number_of_reviews and reviews_per_month, calculated_host_listings_count and availability_365

Negative corr : price and longitude, number_of_reviews and id, reviews_per_month and minimum_nights
